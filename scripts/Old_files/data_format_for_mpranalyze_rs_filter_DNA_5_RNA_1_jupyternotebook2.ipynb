```python
dnacount = df2_dropped[["rsID_UMI", "DNArep1", "DNArep2", "DNArep3"]]
rnacount = df2_dropped[["rsID_UMI", "RNArep1", "RNArep2", "RNArep3"]]
```

```python
dnacount_rep1 = dnacount[["rsID_UMI","DNArep1"]]
dnacount_rep2 = dnacount[["rsID_UMI","DNArep2"]]
dnacount_rep3 = dnacount[["rsID_UMI","DNArep3"]]
rnacount_rep1 = rnacount[["rsID_UMI","RNArep1"]]
rnacount_rep2 = rnacount[["rsID_UMI","RNArep2"]]
rnacount_rep3 = rnacount[["rsID_UMI","RNArep3"]]
```

```python
#dnacount_rep1["rep"]= "rep1" added a column with rep1 rep2 or rep3 for identification 
dnacount_rep1.loc[:, 'rep'] = 'rep1'
dnacount_rep2.loc[:, 'rep'] = 'rep2'
dnacount_rep3.loc[:, 'rep'] = 'rep3'
rnacount_rep1.loc[:, 'rep'] = 'rep1'
rnacount_rep2.loc[:, 'rep'] = 'rep2'
rnacount_rep3.loc[:, 'rep'] = 'rep3'
```

```python
dnacount_rep1.columns= ["rsID_UMI", "count", "rep"]
dnacount_rep2.columns= ["rsID_UMI", "count", "rep"]
dnacount_rep3.columns= ["rsID_UMI", "count", "rep"]
rnacount_rep1.columns= ["rsID_UMI", "count", "rep"]
rnacount_rep2.columns= ["rsID_UMI", "count", "rep"]
rnacount_rep3.columns= ["rsID_UMI", "count", "rep"]
```

```python
all_dnacount = [dnacount_rep1, dnacount_rep2, dnacount_rep3]
all_rnacount = [rnacount_rep1, rnacount_rep2, rnacount_rep3]

```

```python
dnacountall = pd.concat(all_dnacount).reset_index(drop=True)
rnacountall = pd.concat(all_rnacount).reset_index(drop=True)
```

```python
dnacountall.columns = ["rsID_UMI", "count", "batch"]
rnacountall.columns = ["rsID_UMI", "count", "batch"]
```

```python
dnacountall[["rsID", "barcode"]] = dnacountall.rsID_UMI.str.split(":", expand=True)
rnacountall[["rsID", "barcode"]] = rnacountall.rsID_UMI.str.split(":", expand=True)
```

```python
dnacountall[["SNP", "allele"]] = dnacountall.rsID.str.split("_", expand=True)
rnacountall[["SNP", "allele"]] = rnacountall.rsID.str.split("_", expand=True)
```

```python
dnacountall["barcode_rep"] = dnacountall["rsID_UMI"]+","+dnacountall["batch"]
rnacountall["barcode_rep"] = rnacountall["rsID_UMI"]+","+dnacountall["batch"]
```
